### Metricas modelo Random Forests

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
from db_connection import read_data
```

```python
df = read_data()
```

```python
df.head()
```

```python
mse = mean_squared_error(df['Happiness_Score'], df['Predicted_Happiness_Score'])
mae = mean_absolute_error(df['Happiness_Score'], df['Predicted_Happiness_Score'])
r2 = r2_score(df['Happiness_Score'], df['Predicted_Happiness_Score'])


print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Happiness_Score', y='Predicted_Happiness_Score', data=df, color='purple', alpha=0.6, edgecolor=None)
plt.plot([df['Happiness_Score'].min(), df['Happiness_Score'].max()], 
         [df['Happiness_Score'].min(), df['Happiness_Score'].max()], 
         'r--', lw=2)
plt.xlabel('Happiness Real Value Score')
plt.ylabel('Happiness Score Predicted Value')
plt.title('Comparison of Actual vs. Predicted Values of Happiness Score')

plt.text(0.05, 0.95, f'MSE: {mse:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.text(0.05, 0.90, f'MAE: {mae:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.text(0.05, 0.85, f'R²: {r2:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='id', y='Happiness_Score', label='Observed')
sns.lineplot(data=df, x='id', y='Predicted_Happiness_Score', label='Estimated')
plt.ylabel('Happiness Score')
plt.title('Observed and Estimated Happiness Scores')
plt.legend()
plt.show()
```
