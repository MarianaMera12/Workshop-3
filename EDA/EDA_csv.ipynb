### Null df values

We can verify that we do not have null values in the dataframe, this is very important when training the model.

```python
dataframes = [df_2015, df_2016, df_2017, df_2018, df_2019]

def total_nulos(dataframes):
    total = []
    for df in dataframes:
        total_nulos_df = df.isnull().sum().sum()
        total.append(total_nulos_df)
    return total

total_nulos_por_df = total_nulos(dataframes)

for i, total in enumerate(total_nulos_por_df, start=2015):
    print(f"Total null values in df_{i}: {total}")
```

We can see that more information is available in other fields of the dataset

```python
def num_columns(dataframes):
    num_columns_list = []
    for df in dataframes:
        num_columns_df = df.shape[1]  
        num_columns_list.append(num_columns_df)
    return num_columns_list

num_columns_per_df = num_columns(dataframes)

for i, num_columns in enumerate(num_columns_per_df, start=2015):
    print(f"Number of columns in df_{i}: {num_columns}")
```

### Standardize Column names 

To ensure a successful merge of the 5 CSV files, we need to standardize the columns. We will start by standardizing the files corresponding to the years 2015 and 2016, as they share a similar column name structure. 

```python
def standardize_column_1516(df):

    new_column_names = {
        'Country': 'Country',
        'Region': 'Region',
        'Happiness Rank': 'Happiness_Rank',
        'Happiness Score': 'Happiness_Score',
        'Standard Error': 'Standard_Error',
        'Economy (GDP per Capita)': 'Economy_GDP_per_Capita',
        'Family': 'Social_Support',
        'Health (Life Expectancy)': 'Health_Life_Expectancy',
        'Freedom': 'Freedom',
        'Trust (Government Corruption)': 'Corruption',
        'Generosity': 'Generosity',
        'Dystopia Residual': 'Dystopia_Residual'
    }

    df.rename(columns=new_column_names, inplace=True)
```

### Model training

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

The features 'Economy_GDP_per_Capita', 'Social_Support', 'Health_Life_Expectancy', 'Freedom', 'Corruption' and 'Generosity', previously identified as the most relevant for predicting the happiness score, were selected. The dataset was then divided into training and test sets, with 70% of the data being allocated to training and 30% to testing.

```python
X = df[['Economy_GDP_per_Capita', 'Social_Support', 'Health_Life_Expectancy', 'Freedom', 'Corruption', 'Generosity']]
y = df['Happiness_Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=59)
```

### Linear Regression Model

We use linear regression as a method to model the relationship between a dependent variable, such as the happiness score, and independent variables, such as selected characteristics. Its goal is to find a straight line that best fits the data to predict the happiness score.

```python
model = LinearRegression()  
model.fit(X_train, y_train)  

y_pred = model.predict(X_test)  

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2:", r2)
```

### Random Forest Model

We used the Random Forest model to predict the happiness score. It is a supervised learning algorithm used for regression tasks. It consists of building multiple decision trees during training and combining their predictions to obtain a final prediction.

```python
random_forest_model = RandomForestRegressor(n_estimators=100, random_state=87)

random_forest_model.fit(X_train, y_train)

y_pred_rf = random_forest_model.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Regression Model Results:")
print("Mean Squared Error (MSE):", mse_rf)
print("Coefficient of determination (R^2):", r2_rf)
```

 Comparing these metrics with those of the Linear Regression model, we can conclude that the Random Forest model performs better in predicting the happiness score, explaining about 85.5% of the variability observed in the data.

```python
importances = random_forest_model.feature_importances_

sorted_indices = importances.argsort()[::-1]
sorted_importances = importances[sorted_indices]
sorted_features = X_train.columns[sorted_indices]

plt.figure(figsize=(10, 6))
plt.bar(range(X_train.shape[1]), sorted_importances, tick_label=sorted_features)
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Importance of Features in Random Forest Regression')
plt.show()
```
