Para conocer un poco sobre las columnas: 

- **Country:** Nombre del país que está siendo evaluado.

- **Region:** La región geográfica a la que pertenece el país.

- **Happiness Rank:** La posición del país en el ranking de felicidad, donde un rango más bajo indica un mayor nivel de felicidad.

- **Happiness Score:** Puntaje de felicidad basado en una escala, generalmente derivado de encuestas sobre la percepción del bienestar y la calidad de vida.

- **Standard Error:** Error estándar del puntaje de felicidad, que indica la precisión de la estimación del puntaje de felicidad.

- **Economy (GDP per Capita):** El Producto Interno Bruto (PIB) per cápita, ajustado por paridad de poder adquisitivo, que refleja la riqueza promedio por persona en el país.

- **Family:** Medida de apoyo social percibido, usualmente basada en la pregunta de si los encuestados tienen a alguien en quien confiar en tiempos de necesidad.

- **Health (Life Expectancy):** Expectativa de vida saludable al nacer, indicando la cantidad de años que una persona puede esperar vivir en buena salud.

- **Freedom:** Medida de la libertad percibida para tomar decisiones importantes en la vida, basada en respuestas a encuestas.
Trust (Government Corruption): Medida de la percepción de la corrupción en el gobierno y las instituciones públicas.

- **Generosity:** Medida de la generosidad percibida, generalmente basada en la cantidad de donaciones recientes hechas a organizaciones benéficas.

- **Dystopia Residual:** Valor calculado que sirve como referencia hipotética para los peores niveles posibles de cada una de las métricas de bienestar. Se utiliza para medir el desvío de un país con respecto a esta referencia negativa.

### Valores nulos de los df

Podemos verirficar que no contamos con valores nulos en los dataframe, esto es muy importante al momento de realizar el entrenamiento del modelo

```python
dataframes = [df_2015, df_2016, df_2017, df_2018, df_2019]

def total_nulos(dataframes):
    total = []
    for df in dataframes:
        total_nulos_df = df.isnull().sum().sum()
        total.append(total_nulos_df)
    return total

total_nulos_por_df = total_nulos(dataframes)

for i, total in enumerate(total_nulos_por_df, start=2015):
    print(f"Total null values in df_{i}: {total}")
```

### Entrenamiento del modelo

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X = df[['Economy_GDP_per_Capita', 'Social_Support', 'Health_Life_Expectancy', 'Freedom', 'Corruption', 'Generosity']]
y = df['Happiness_Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=59)
```

### Modelo Regresion Lineal

```python
model = LinearRegression()  
model.fit(X_train, y_train)  

y_pred = model.predict(X_test)  

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2:", r2)
```

### Modelo Random Forest

```python
random_forest_model = RandomForestRegressor(n_estimators=100, random_state=87)

random_forest_model.fit(X_train, y_train)

y_pred_rf = random_forest_model.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Regression Model Results:")
print("Mean Squared Error (MSE):", mse_rf)
print("Coefficient of determination (R^2):", r2_rf)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_rf, color='purple', edgecolor='purple', s=50, alpha=0.7, label='Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', linewidth=2, label='Ideal Line')
plt.xlabel('Actual Values', fontsize=12)
plt.ylabel('Predicted Values', fontsize=12)
plt.title('Predictions vs. Actual Values', fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
importances = random_forest_model.feature_importances_

sorted_indices = importances.argsort()[::-1]
sorted_importances = importances[sorted_indices]
sorted_features = X_train.columns[sorted_indices]

plt.figure(figsize=(10, 6))
plt.bar(range(X_train.shape[1]), sorted_importances, tick_label=sorted_features)
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Importance of Features in Random Forest Regression')
plt.show()
```
